# <span style="color:Green">P4: How many markets operate year-round? What is the distribution of market visits by month?</span>

```python
import pandas as pd

# Read the Excel dataset, ignore any index column
farmer_market = pd.read_excel("Farmers' Market Observation-dataset.xlsx", index_col=None)
# Find all entries that have Yearround value of "Yes", then count the number of those entries based on market_ID column
yearround_market_amount = farmer_market[farmer_market.Yearround == "Yes"].market_ID.count()

# Group the dataset on Visit_month column, then count all entries in each group
# This will be the total visits for each month
visit_dist_by_month = farmer_market.groupby("Visit_month").market_ID.count()

print(f"Number of markets that operate year-round: {yearround_market_amount} markets")
print(f"Distribution of market visits by month")
print(visit_dist_by_month)
print("Based on the total visits of each month, the market visits mostly fall between Summer and "
      "beginning of Autumn, with the highest number of visits in August")
```

# <span style="color:Green">P5: How many markets have communal sinks available? How many markets have pet signs?</span>

```python
# Find all entries that have communal_sinks value of "Yes", then count them based on market_ID
communal_sink_available = farmer_market[farmer_market.communal_sinks == "Yes"].market_ID.count()
# Find all entries that have pet_signs value of "Yes", then count them based on market_ID
pet_sign_available = farmer_market[farmer_market.pet_signs == "Yes"].market_ID.count()

print(f"Number of markets that have communal sinks: {communal_sink_available:5} markets")
print(f"Number of markets that have pet signs: {pet_sign_available:10} markets")
```

# <span style="color:Green">P6: What percentage of vendors sell raw meat (TCS_rawmeat)? What percentage of vendors handle ready-to-eat meats (TCS_rtemeat)? </span>

```python
# Count all entries in market_ID, this will be the total markets
total_market = farmer_market.market_ID.count()

# Find all entries that have TCS_rawmeat value of 1, then count them based on market_ID
raw_meat_vendor = farmer_market[farmer_market.TCS_rawmeat == 1].market_ID.count()
# Calculate the percentage by dividing number of raw meat vendors with total market count, then multiply by 100
raw_meat_percentage = (raw_meat_vendor / total_market) * 100

# Find all entries that have TCS_rtemeat value of 1, then count them based on market_ID
rte_meat_vendor = farmer_market[farmer_market.TCS_rtemeat == 1].market_ID.count()
# Calculate the percentage by dividing number of RTE meat vendors with total market count, then multiply by 100
rte_meat_percentage = (rte_meat_vendor / total_market) * 100

print(f"Percentage of vendors sell raw meat: {raw_meat_percentage:20.2f}%")
print(f"Percentage of vendors sell ready-to-eat meat: {rte_meat_percentage:11.2f}%")
```

```python
# Table 3
# beh_cash - Handle cash before food
# beh_handle_food - Handle raw meat before rte food
# beh_touchface - Touch mouth, hair, or face
# beh_eatdrink - Eat or drink
# beh_phone - Handle phone
# beh_belongings - Handle other personal belongings
# beh_waste - Handle waste
# beh_dishes - Handle dirty dishes
# beh_pickup - Pickup something on the ground
# beh_coughsneeze - Cough, sneeze, or blow nose
# beh_smoke - Smoke
# beh_animals - Touch animals

# Clean dataset by removing rows that have NaN values in the specified columns below (only row 325 was removed)
farmer_market_clean = farmer_market.dropna(subset=[
    "beh_cash", "beh_handle_food", "beh_touchface", "beh_eatdrink", "beh_phone", "beh_belongings", "beh_waste",
    "beh_dishes", "beh_pickup", "beh_coughsneeze", "beh_smoke", "beh_animals"
    ])

# Extract required columns out
beh_cash = farmer_market_clean.beh_cash.astype(int)
beh_touchface = farmer_market_clean.beh_touchface.astype(int)
beh_eatdrink = farmer_market_clean.beh_eatdrink.astype(int)
beh_phone = farmer_market_clean.beh_phone.astype(int)
beh_handle_food = farmer_market_clean.beh_handle_food.astype(int)
beh_belongings = farmer_market_clean.beh_belongings.astype(int)
beh_waste = farmer_market_clean.beh_waste.astype(int)
beh_dishes = farmer_market_clean.beh_dishes.astype(int)
beh_pickup = farmer_market_clean.beh_pickup.astype(int)
beh_coughsneeze = farmer_market_clean.beh_coughsneeze.astype(int)
beh_smoke = farmer_market_clean.beh_smoke.astype(int)
beh_animals = farmer_market_clean.beh_animals.astype(int)

# Concatenate extracted columns into a DataFrame
extracted_df = pd.concat([
    beh_cash, beh_touchface, beh_eatdrink, beh_phone, beh_handle_food, beh_belongings, beh_waste, 
    beh_dishes, beh_pickup, beh_coughsneeze, beh_smoke, beh_animals
    ], axis=1)

# Calculate the total behaviours for each row
beh_total = extracted_df.sum(axis=1)
# Assign the total value to a new column
extracted_df["beh_total"] = beh_total
# Calculate mean and standard deviation for each column
mean = extracted_df.mean(skipna=True).round(3)
std = extracted_df.std().round(3)
# Form the range by finding min and max value for each column
the_range = pd.Series(
    {
        "beh_cash": f"{beh_cash.min()} - {beh_cash.max()}",
        "beh_touchface": f"{beh_touchface.min()} - {beh_touchface.max()}",
        "beh_eatdrink": f"{beh_eatdrink.min()} - {beh_eatdrink.max()}",
        "beh_phone": f"{beh_phone.min()} - {beh_phone.max()}",
        "beh_handle_food": f"{beh_handle_food.min()} - {beh_handle_food.max()}",
        "beh_belongings": f"{beh_belongings.min()} - {beh_belongings.max()}",
        "beh_waste": f"{beh_waste.min()} - {beh_waste.max()}",
        "beh_dishes": f"{beh_dishes.min()} - {beh_dishes.max()}",
        "beh_pickup": f"{beh_pickup.min()} - {beh_pickup.max()}",
        "beh_coughsneeze": f"{beh_coughsneeze.min()} - {beh_coughsneeze.max()}",
        "beh_smoke": f"{beh_smoke.min()} - {beh_smoke.max()}",
        "beh_animals": f"{beh_animals.min()} - {beh_animals.max()}",
        "beh_total": f"{beh_total.min()} - {beh_total.max()}"
    }
)

# Create a DataFrame from the calculated mean, std, and range above
table_3 = pd.DataFrame(
        {"Mean frequency observed per vendor": mean, "SD": std, "Range": the_range}
    )
# Rename column names with the correct ones from the paper
table_3 = table_3.rename(index={
        "beh_cash": "Handle cash before handling food again", "beh_touchface": "Touch mouth, hair, or face", 
        "beh_eatdrink": "Eat or drink", "beh_handle_food": "Handle a raw food before a ready-to-eat food",
        "beh_phone": "Handle phone", "beh_belongings": "Handle other personal belongings",
        "beh_waste": "Handle waste container or garbage", "beh_dishes": "Handle dirty dishes or utensils", 
        "beh_pickup": "Pickup something from the ground", "beh_coughsneeze": "Cough, sneeze, or blow nose",
        "beh_smoke": "Smoke", "beh_animals": "Touch animals", "beh_total": "Total behaviours requiring handwashing"
        }
        )
# Rename the index column with the correct one from the paper
table_3.index.name = "Behaviours"
# Print the DataFrame out
table_3
```
